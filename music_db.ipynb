#Dependencies

```python
import pandas as pd
```

#Spotify's Top 100 2017 Dataset

```python
#Import Spotify DB
spotify_df=pd.read_csv("Resources/featuresdf.csv", encoding="utf-8")
spotify_df.head()
```

#MTV's Top Music Artists Dataset (w/Facebook, Twitter, and Website info)

```python
#Import media DB
social_df=pd.read_csv("Resources/Artist_socialmedia.csv", encoding="utf-8")
social_df.head()
```

#Data cleanup

```python
#Clean up social media df including facebook, twitter, and website info with artist name. Rename "name" to "artists" 
#to match Spotify df
new_social_df = social_df[["name","facebook","twitter","website"]].copy()
new_social_df=new_social_df.rename(columns={"name":"artists"})
new_social_df.head()
```

```python
#Reduce Spotify df including Spotify song_id, song name, and artist. Rename "id" to "Spotify_id" to better indentify column
new_spotify_df = spotify_df[['id', 'name', 'artists']].copy()
new_spotify_df=new_spotify_df.rename(columns={"id":"Spotify_id"})
new_spotify_df.head()
```

```python
#Drop Duplicates in Spotify DB
spotify_drop = new_spotify_df.drop_duplicates('artists')
spotify_drop.head()
```

```python
#Merge the two reduced df's. Change order or columns for better reading
merged_df = pd.merge(spotify_drop, new_social_df, on="artists",how="left")
merged_df = merged_df[["artists", "name", "Spotify_id", "website", "facebook", "twitter"]]
merged_df.head()
```

MongoDB

```python
import pymongo
```

```python
#Establish connection
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
#Create schema
musicDB = client["musicDataBase"]
#Drop artist collection
musicDB.artists.drop()
#Create collection
artist_col = musicDB["artists"]
```

```python
#Insert records into collection "artists"
xds = artist_col.insert_many(merged_df.to_dict('records'))
print(xds.inserted_ids)
```
