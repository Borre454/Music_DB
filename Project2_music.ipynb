```python
spotify_df=pd.read_csv("featuresdf.csv")
spotify_df.head()
```

```python
social_df=pd.read_csv("Artist_socialmedia.csv")
social_df.head()
```

```python
#clean up social media df for just facebook and twitter with artist name
new_social_df = social_df[["name","facebook","twitter","website"]].copy()
new_social_df=new_social_df.rename(columns={"name":"artists"})
new_social_df.head()
```

```python
new_spotify_df = spotify_df[['id', 'name', 'artists']].copy()
new_spotify_df=new_spotify_df.rename(columns={"id":"Spotify_id"})
new_spotify_df.head()
```

```python
spotify_drop = new_spotify_df.drop_duplicates('artists')
spotify_drop.head()
```

```python
pd.merge(spotify_drop, new_social_df, on="artists",how="left")
#new_spotify_df.merge(new_social_df, )
```
